# 1. Vector Database 이해

## 1.1 Vector Database란?
벡터 임베딩 또는 텍스트, 이미지, 오디오와 같은 구조화되지 않은 데이터의 수치 표현을 저장, 색인 생성, 쿼리할 수 있게 해주는 모든 데이터베이스

## 1.2 vector embedding이란?
image.png

수치화된 데이터를 다차원 공간에 표현하고 <br>
길이, 방향, 각도를 비교하여 비슷한 데이터를 유사성이 높다고 표현하는 방법
